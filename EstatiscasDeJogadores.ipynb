```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colunas_desejadas = ["Player","Pos", "Squad","Comp", "Gls", "Ast", "G+A", "Saves", "Tkl+Int", "KP"]

df = pd.read_csv(r"/players_data_light-2024_2025.csv", usecols=colunas_desejadas)

df
```

```python
print("\nJogadores por liga:")
print(df["Comp"].value_counts())

print("\nDistribuição de posições:")
print(df["Pos"].value_counts())
```

```python
s = "DF, MF"
s.split(",")[0]
```

```python
df["Pos"] = df ["Pos"].apply(lambda pos: pos.split(",")[0])
print("\nDistribuição de posições:")
print(df["Pos"].value_counts())
```

```python
top_scorers = df.sort_values(by="Gls", ascending=False).head(10)[
    ["Player", "Squad", "Comp", "Pos", "Gls"]
]
print(top_scorers)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_scorers, x="Gls", y="Player", hue="Comp")
plt.xlabel("Gols")
plt.ylabel("Jogador")
plt.title("Top 10 Jogadores por Gols")
plt.legend(title="Liga")
plt.tight_layout()

plt.show()
```

```python
#Melhores em Participação em Gols e ASS
top_ofensivos = df.sort_values(by="Gls", ascending=False).head(10)[
    ["Player", "Squad", "Comp", "Pos", "G+A"]
]
print(top_ofensivos)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_ofensivos, x="G+A", y="Player", hue="Comp")
plt.ylabel("Jogador")
plt.title("Top 10 Jogadores por Gols e Assistencias")
plt.legend(title="Liga")
plt.tight_layout()

plt.show()
```

```python
#Melhores em Interciptações e Desarmes
top_defensivos = df.sort_values(by="Tkl+Int", ascending=False).head(10)[
    ["Player", "Squad", "Comp", "Pos", "Tkl+Int"]
]
print(top_scorers)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_defensivos, x="Tkl+Int", y="Player", hue="Comp")
plt.ylabel("Jogador")
plt.title("Top 10 Jogadores com mais desarmes")
plt.legend(title="Liga")
plt.tight_layout()

plt.show()
```

```python
#Melhor Liga com qualidade Defensiva
df.groupby("Comp")["Tkl+Int"].mean().sort_values(ascending=False)
```

```python
#Melhor Liga em qualidade Ofensiva
df.groupby("Comp")["G+A"].mean().sort_values(ascending=False)
```

```python
top_goleiros = df.sort_values(by="Saves", ascending=False).head(10)[
    ["Player", "Squad", "Comp", "Pos", "Saves"]
]
print(top_goleiros)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_goleiros, x="Saves", y="Player", hue="Comp")
plt.ylabel("Jogador")
plt.title("Top 10 goleiros com mais defesas")
plt.legend(title="Liga")
plt.tight_layout()

plt.show()
```

**MELHOR TIME DO ANO**




```python
goleiros = df[df["Pos"] == "GK"]
defensivos = df[df["Pos"] == "DF"]
meias = df[df["Pos"] == "MF"]
atacantes = df[df["Pos"] == "FW"]
```

```python
melhor_goleiro = goleiros.sort_values(by="Saves", ascending=False).head(1)[
    ["Player", "Squad", "Comp", "Pos"]
]
display(melhor_goleiro)


melhor_defensor = defensivos.sort_values(by="Tkl+Int", ascending=False).head(3)[
    ["Player", "Squad", "Comp", "Pos"]
]
display(melhor_defensor)

melhor_meia = meias.sort_values(by="KP", ascending=False).head(5)[
    ["Player", "Squad", "Comp", "Pos"]
]
display(melhor_meia)

melhor_atacante = atacantes.sort_values(by="G+A", ascending=False).head(2)[
    ["Player", "Squad", "Comp", "Pos"]
]
display(melhor_atacante)
```

```python
dream_team = pd.concat(
    [melhor_goleiro, melhor_defensor, melhor_meia, melhor_atacante], ignore_index=True
)
display(dream_team)
```
